```python
import numpy as np
import pandas as pd
import difflib
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
movies_data = pd.read_csv('/content/movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
#selecting only relevant features that  which are usefull for us in recommending the movies

selected_features = ['genres', 'keywords', 'cast', 'director', 'tagline', 'original_title']
print(selected_features)
```

```python
#replacing the missing values or strings in the dataset of our selected features, so we are going to use a for loop
for feature in selected_features:
  movies_data[feature]=movies_data[feature].fillna('')#with this we can only fill the selected features not all.
```

```python
movies_data.isnull().sum()
```

```python
#combinig all the slected features
combined_features = movies_data['genres']+' '+movies_data['original_title']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
#converting the text data into numerical values
vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
#finding the similarity using Cosine similarity
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
#getting the User data
movie_name = input('Enter YOUR (MovieName/Cast/Director/Tagline/Keywords/Genre):')
```

```python
#creating a list eith all movie names given in the dataset
list_of_all_titles = movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
#finding the close match for the movie name given by the user
find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with the title
index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]
print(index_of_movie)
```

```python
#getting a list of similar movie
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorted the movies based on their similarity score
sorted_similarity_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)
print(sorted_similarity_movies)
```

```python
#printing the similar movies based on the index
print('movies suggested for you:\n')
i=1
for movie in sorted_similarity_movies:
  index=movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]

  if(i<30):
    print(i,'_', title_from_index)
    i+=1
```

Movie Recommendation System


```python
movie_name = input('Enter YOUR (MovieName/Cast/Director/Tagline/Keywords/Genre):')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_movie]))

sorted_similarity_movies = sorted(similarity_score, key=lambda x:x[1], reverse=True)

print('Movies suggested for you:\n')
i=1
for movie in sorted_similarity_movies:
  index=movie[0]
  title_from_index = movies_data[movies_data.index==index]['title'].values[0]

  if(i<30):
    print(i,'_', title_from_index)
    i+=1
```

```python
import pickle

# Save the movie data and similarity
pickle.dump(movies_data, open('movies.pkl', 'wb'))
pickle.dump(similarity, open('similarity.pkl', 'wb'))
```
